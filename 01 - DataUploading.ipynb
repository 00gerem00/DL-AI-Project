```python
import os
from google.colab import drive

# 1. Monta Drive
drive.mount('/content/drive')

# 2. Vai nella cartella del progetto
# Ricordati che Lorenzo dovr√† cambiare questo percorso con il suo!
os.chdir('/content/drive/MyDrive/00gerem00/DL-AI-Project')

# 3. Indica a Python dove si trova il passaporto kaggle.json
os.environ['KAGGLE_CONFIG_DIR'] = "/content/drive/MyDrive"

# 4. Scarica il dataset specifico
# Vai sulla pagina Kaggle del dataset, clicca sui tre puntini e "Copy API command"
# Incolla qui solo la parte finale (es: 'utente/nome-dataset')
dataset_api_name = 'praveen2084/leather-defect-classification'

!kaggle datasets download -d {dataset_api_name}

# 5. Estrai i dati in una cartella chiamata 'dataset'
if not os.path.exists('dataset'):
    !unzip -q *.zip -d dataset
    !rm *.zip
    print("Dataset scaricato ed estratto con successo!")
else:
    print("La cartella dataset esiste gi√†.")
```

```python
import os

# 1. Trova la cartella intermedia automaticamente
base_dir = 'dataset'
contenuto_base = os.listdir(base_dir)

# Cerchiamo la cartella che contiene le classi (es. "Leather Defect Classification")
# Prendiamo la prima cartella che troviamo
folder_intermedia = [f for f in contenuto_base if os.path.isdir(os.path.join(base_dir, f))][0]
data_dir = os.path.join(base_dir, folder_intermedia)

print(f"üìÇ Percorso completo trovato: {data_dir}")

# 2. Conta le immagini per ogni classe
print("\nüìä Statistiche del Dataset:")
totale_immagini = 0

for classe in os.listdir(data_dir):
    path_classe = os.path.join(data_dir, classe)
    if os.path.isdir(path_classe):
        n_img = len(os.listdir(path_classe))
        totale_immagini += n_img
        print(f"   -> Classe '{classe}': {n_img} immagini")

print(f"\n‚úÖ Totale immagini trovate: {totale_immagini}")
```
