In the following *notebook* we upload from *www.kaggle.com* the dataset on which our computer vision project will be carried on.

```python
import os
from google.colab import drive

drive.mount('/content/drive')

path_MG = '/content/drive/MyDrive/00gerem00/DL-AI-Project'
path_LM = '/content/drive/MyDrive/lorenzomeroni02/DL-AI-Project'

if os.path.exists(path_MG):
    print(f"Enter in: {path_MG}")
    os.chdir(path_MG)
elif os.path.exists(path_LM):
    print(f"Enter in: {path_LM}")
    os.chdir(path_LM)
else:
    print("Error!")
    print("Check if you created the folder in your Drive!")

os.environ['KAGGLE_CONFIG_DIR'] = "/content/drive/MyDrive"
```

```python
dataset_api_name = 'praveen2084/leather-defect-classification'
zip_filename = 'leather-defect-classification.zip'

# Check if we already downloaded the ZIP in our Drive
if not os.path.exists(zip_filename):
    print("ZIP not found on Drive! Starting the download from Kaggle...")
    !kaggle datasets download -d {dataset_api_name}
    print("Download completed! ZIP file safely stored on Drive.")
else:
    print("The ZIP file already exists on Drive! No need to download it again.")
```

```python
import shutil

# Define the fast local directory on Colab
fast_local_dir = '/content/fast_dataset'

# Clean up if it already exists from a previous run
if os.path.exists(fast_local_dir):
    shutil.rmtree(fast_local_dir)
os.makedirs(fast_local_dir, exist_ok=True)

print("Extracting the ZIP file...")
# Extract from the ZIP on Drive directly to the /content/ folder
!unzip -q {zip_filename} -d {fast_local_dir}
print("Extraction completed!")
```

```python
base_dir = '/content/fast_dataset'
base_content = os.listdir(base_dir)

# Find the subfolder containing the classes
intermediate_folder = [f for f in base_content if os.path.isdir(os.path.join(base_dir, f))][0]
data_dir = os.path.join(base_dir, intermediate_folder)

print(f"Dataset: {data_dir}")

# Count images for each class
print("\n Dataset Statistics:")
total_images = 0

for class_name in os.listdir(data_dir):
    class_path = os.path.join(data_dir, class_name)
    if os.path.isdir(class_path):
        n_img = len(os.listdir(class_path))
        total_images += n_img
        print(f"    {class_name}: {n_img} images")

print(f"\n Total images found: {total_images}")
```
